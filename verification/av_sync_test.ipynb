```python
from glob import glob
import os
```

```python
experiment_no = 496
file_idx = 19
area = 'burrow' # options are {"nest","center","gily_center","burrow"} - area of arena to check

path = f"Y:/big_setup/gily_experiments/experiment_{experiment_no}/concatenated_data_cam_mic_sync/"
sync_output = f"Y:/big_setup/gily_experiments/experiment_{experiment_no}/sync_test/"
os.makedirs(sync_output, exist_ok=True)

channel_dict = {"nest":[0],"burrow":[1],"center":[2],"gily_center":[4]}
vid_dict = {'center':'center',"nest":"nest_top","burrow":"burrow_side","gily_center":"gily_center"}

try:
    channels = channel_dict[area]
except:
    raise Exception("invalid area input")

audio = None
for ch in channels: 
    str_1 = "%02d"%(ch)
    str_2 = "%03d"%(file_idx)
    # getting the files
    audio_file = glob(path+f"channel_{str_1}_file_{str_2}.wav")[0]

str_1 = vid_dict[area]
str_2 = "%03d"%(file_idx)
# getting the video file
video_file = glob(path+f"video_{str_1}_{str_2}.mp4")[0]
```

### For FFMPEG, use the miniforge env ffmpeg-env

```python
import subprocess
output_path = f"{sync_output}av_{area}_file_{str_2}.mp4"

subprocess.run([
    "ffmpeg",
    "-i", video_file,
    "-i", audio_file,
    "-c:v", "copy",
    "-c:a", "aac",
    output_path
])
```
