```python
import glob
import os
```

```python
experiments = range(500,568)
```

### To check whether timestamps are run

```python
for exp in experiments:
    print(f"For experiment {exp}")
    path = f"/mnt/home/njoseph/saneslab_data/big_setup/experiment_{exp}/"
    length_logs = glob.glob(path+"length_log*.txt")
    for i in length_logs:
        with open(i, "r") as f:
            lines = f.readlines()
            for line in lines[-4:]:
                print(line, end="")
    print("================================================================================================================\n\n\n")    
    
```

### To check whether concat is complete

```python
for exp in experiments:
    print(f"For experiment {exp}")
    path = f"/mnt/home/njoseph/saneslab_data/big_setup/experiment_{exp}/concatenated_data_cam_mic_sync/"
    wav_files = len(glob.glob(path+"*.wav"))
    cam_files = len(glob.glob(path+"*mp4"))
    no_cam_file_idx = cam_files/7
    no_wav_file_idx = wav_files/21
    if exp >=515:
        no_wav_file_idx = wav_files/14
    print(f"No wav file indices:{no_wav_file_idx}")
    print(f"No camera file indices:{no_cam_file_idx}")
```

### testing specific experiments

```python
for exp in [548]:
    print(f"For experiment {exp}")
    path = f"/mnt/home/njoseph/saneslab_data/big_setup/experiment_{exp}/concatenated_data_cam_mic_sync/"
    # Getimport glob

    files = glob.glob(path+"channel_*_file_*.wav")
    
    channels = set()
    for f in files:
        parts = os.path.basename(f).split("_")
        channel = parts[1]  # acquisition_data_<channel>_<file>.wav
        channels.add(channel)
    channels = list(channels)
    channels.sort()
    print("Number of channels:", len(channels))

    for ch in channels:
        files = glob.glob(path+f"channel_{ch}_file_*.wav")
        files.sort()
        print(f"Number of files in audio channel {ch}: {len(files)}")
        print(f"Last file in audio channel {ch}: {files[-1]} \n\n")

    cameras = set()
    files = glob.glob(path+"video_*.mp4")
    for f in files:
        parts = os.path.basename(f).split("_")
        cam = "_".join(parts[1:-1])   # acquisition_data_<channel>_<file>.wav
        cameras.add(cam)
    
    print("Number of cameras:", len(cameras))


    for cam in cameras:
        files = glob.glob(path+f"video_{cam}_*.mp4")
        files.sort()
        print(f"Number of files for camera {cam}: {len(files)}")
        print(f"Last file for camera {cam}: {files[-1]} \n\n")
```
